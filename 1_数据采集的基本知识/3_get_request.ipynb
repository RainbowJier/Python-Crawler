# 1、什么是get请求
    GET方法是最常见也是最简单的，HTTP默认的请求方法就是GET。一般用于我们向服务器获取数据，可以直接将URL输入，不需要其它的转换，即所有需要请求的信息都包含在URL中。

```python
'''get请求'''
import requests

url="https://www.baidu.com/"  #请求目标链接
response=requests.get(url=url) #获取相应内容
print(response) #返回200表示成功获取数据
```

```python
import  requests
from bs4 import BeautifulSoup  #用于解析网页

url="https://www.ahpu.edu.cn/"
head = {
    'User-agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/60.0.3112.113 Safari/537.36'
}
r=requests.get(url=url,headers=head)
if r.status_code==200:       #判读服务器时候请求成功
    r.encoding=r.apparent_encoding  #设置网页格式
    print(r.text)   #返回网页源代码
    
```
